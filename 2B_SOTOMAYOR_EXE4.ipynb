# **Importing Libraries**

```python
import pandas as pd #Data manipulation
import numpy as np #Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression #Data modeling
from sklearn.preprocessing import StandardScaler
```

# **1. Data Exploration:**

```python
data_path = "Lesson_4_student_grades_dataset.csv"
dataset = pd.read_csv(data_path)
```

***As we can see in the scatter plot the 5 variables has an effect to student's Grades.***

```python
plt.scatter(dataset['Hours_of_Study'], dataset['Grades'])
plt.xlabel('Independent Variable (X)')
plt.xlabel('Dependent Variable (Y)')
plt.title("Student's Study Hours")
plt.show()

plt.scatter(dataset['Attendance_Percentage'], dataset['Grades'])
plt.xlabel('Independent Variable (X)')
plt.xlabel('Dependent Variable (Y)')
plt.title("Student's Attendance")
plt.show()

plt.scatter(dataset['Previous_Exam_Scores'], dataset['Grades'])
plt.xlabel('Independent Variable (X)')
plt.xlabel('Dependent Variable (Y)')
plt.title("Student's Exam Scores")
plt.show()

plt.scatter(dataset['Extracurricular_Activities'], dataset['Grades'])
plt.xlabel('Independent Variable (X)')
plt.xlabel('Dependent Variable (Y)')
plt.title("Student's Extracurricular Activities")
plt.show()

plt.scatter(dataset['Travel_Time_to_School'], dataset['Grades'])
plt.xlabel('Independent Variable (X)')
plt.xlabel('Dependent Variable (Y)')
plt.title("Student's Travel Time")
plt.show()
```

# **2. Data Preprocessing:**

**The importance of normalizing or standardizing numerical features enhances data analysis and modeling accuracy**

```python
X = dataset.drop('Grades', axis=1)
Y = dataset[('Grades')]
```

# **3. Train-Test Split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

# **4. Linear Regression Model:**

```python
model = LinearRegression()
```

# **5. Model Training:**

```python
model.fit(X_train, Y_train)
```

# **6. Prediction and Evaluation**

```python
new_data = np.array([[10, 100, 100, 1, 0]])
prediction = model.predict(new_data)
print(f"Predicted Grade: {prediction[0]}")
```

# **7. Feature Importance Analysis:**

The importance of of using each features to predict the student's grades is massive, because each features have an effect to student's grades

```python
print("Hours_of_Study")
slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)

print("")

print("Attendance_Percentage")
slope = model.coef_[1]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)

print("")

print("Previous_Exam_Scores")
slope = model.coef_[2]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)

print("")

print("Extracurricular Activities")
slope = model.coef_[3]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)

print("")

print("Travel_Time_to_School")
slope = model.coef_[4]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

**The most significant impact on student grades are hours study, attendance percentage, and previous exam scores. The higher the input in these features the higher the student's grades.**

**Travel time to school has a slight impact to student's grades**
